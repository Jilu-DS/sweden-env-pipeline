```python
from dotenv import load_dotenv
import os

load_dotenv()

API_KEY = os.getenv("API_KEY")

print(API_KEY)
```

```python
import requests

city = "Stockholm"

url = f"http://api.openweathermap.org/data/2.5/weather?q={city},se&appid={API_KEY}&units=metric"

response = requests.get(url)

data = response.json()

print(data)
```

```python
temperature = data['main']['temp']
humidity = data['main']['humidity']
weather = data['weather'][0]['description']

print(temperature, humidity, weather)
```

```python
cities = ["Stockholm", "Gothenburg", "Malmo"]

records = []

for city in cities:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city},se&appid={API_KEY}&units=metric"
    data = requests.get(url).json()
    
    record = {
        "city": city,
        "temperature": data['main']['temp'],
        "humidity": data['main']['humidity'],
        "weather": data['weather'][0]['description']
    }
    
    records.append(record)

print(records)
```

```python
import pandas as pd

df = pd.DataFrame(records)

print(df)
```

```python
df.to_csv("data/weather_data.csv", index=False)
```

```python
X = df[['temperature', 'humidity']]
y = df['temperature']   # simple demo
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("Model trained")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['humidity'], df['temperature'])
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.title("Humidity vs Temperature")
plt.show()
```
